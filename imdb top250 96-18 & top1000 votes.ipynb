```python
import pandas as pd
import numpy as np
import re
import lxml

from bs4 import BeautifulSoup
from requests import get
```

# Getting Urls From IMDB's Pollmaster Account

```python
page=get('https://www.imdb.com/user/ur48187336/lists')
soup = BeautifulSoup(page.content, 'lxml')
content = soup.find(id="main")
urls=pd.DataFrame()
temp=pd.DataFrame()

for i in range(len(content.find_all("a", class_="list-name"))):
    link=content.find_all("a", class_="list-name")[i].get('href')
    year=content.find_all("a", class_="list-name")[i].text
    inf=[year, link]
    temp=temp.append(inf)
    urls=urls.append(temp.transpose())
    temp=pd.DataFrame()

urls=urls[urls[0].str.startswith('IMDb Top 250')==True]
urls.columns=['year', 'link']
urls['year']=urls['year'].apply(lambda x: 'imdb_' + str(x[13:]))
urls.reset_index(inplace=True)
urls['link']=urls['link'].apply(lambda x: 'https://www.imdb.com' + x)
urls=urls.sort_values(['year'])
urls=urls.drop(['index'], axis=1)
urls=urls.reset_index()
urls=urls.drop('index', axis=1)
urls['page2']=urls['link'].apply(lambda x: x + "?page=2" )
urls['page3']=urls['link'].apply(lambda x: x + "?page=3" )

urls
```

# Creating A Function For Getting Informations From IMDB List

```python
def imdb_100(url, x):
    df=pd.DataFrame()
    fin=pd.DataFrame()
    
    page = get(url)
    soup = BeautifulSoup(page.content, 'lxml')
    content = soup.find(id="main")

    Frame = content.find_all("div", class_="lister-item mode-detail")

    for i in range(x):
        temp=pd.DataFrame()
        
        articleTitle = soup.find("h1", class_="header").text.replace("\n","")
        FirstLine = Frame[i].find("h3", class_="lister-item-header")

        Ranking = FirstLine.find("span").text
        IMDBlink = FirstLine.find("a").get("href")

        Title = FirstLine.find("a").text
        Date = re.sub(r"[()]","", FirstLine.find_all("span")[-1].text)
        RunTime = Frame[i].find("span", class_="runtime").text[:-4]
        Genre = Frame[i].find("span", class_="genre").text.rstrip().replace("\n","")
        Rating = Frame[i].find("span", class_="ipl-rating-star__rating").text
        try:
            Score = Frame[i].find("span", class_="metascore favorable").text.rstrip()
        except:
            Score = None

        Votes = Frame[i].find_all("p", class_="text-muted")[-1].text.lstrip().replace('\n', '').split('|')[0][6:]

        try:
            Gross = Frame[i].find_all("p", class_="text-muted")[-1].text.lstrip().replace('\n', '').split('|')[1][7:]
        except:
            Gross=None

        a=Frame[i].find_all("p", class_="text-muted text-small")[1].text.split('|')
        Director=a[0].strip().replace('\n', ' ')
        a=Frame[i].find_all("p", class_="text-muted text-small")[1].text.split('|')
        Cast=a[1].strip().replace('\n', ' ')
        lis=[articleTitle, Ranking, IMDBlink, Title, Date, RunTime, Genre, Rating, Score, Votes, Gross, Director, Cast]
        temp=temp.append(lis)
        df=df.append(temp.transpose())
        
    return df
```

# Also, A Function For Correcting Some Infos

```python
def correct(df, x):
    columns=["IMDByear", "Ranking", "IMDBlink", "Title", "Date", "RunTime", "Genre", "Rating", "Score", "Votes", "Gross", "Director", "Cast"]

    df.columns=columns

    df.set_index('Ranking', inplace=True)

    df['Cast1']=df['Cast'].apply(lambda x: x.split(',')[0])
    df['Cast2']=df['Cast'].apply(lambda x: x.split(',')[1])

    df['Cast3']='x'
    for i in range(x):
        try:
            df['Cast3'][i]=df['Cast'][i].split(',')[2]

        except:
            df['Cast3'][i]=None

    df['Cast4']='x'
    for i in range(x):
        try:
            df['Cast4'][i]=df['Cast'][i].split(',')[3]

        except:
            df['Cast4'][i]=None

    df['Votes']=df['Votes'].apply(lambda x: x.replace(',', ''))
    df['Gross']=df['Gross'].apply(lambda x: x.replace('$', '') if x!=None else x)
    df['Gross']=df['Gross'].apply(lambda x: x.replace('M', '') if x!=None else x)
    df=df.apply(pd.to_numeric, errors='ignore')
    
    df['Cast1']=df['Cast1'].apply(lambda x: x.replace('Stars:', ''))
    df['Director']=df['Director'].apply(lambda x: x.replace('Directors: ', ''))
    df['Director']=df['Director'].apply(lambda x: x.replace('Director: ', ''))
    df.drop('Cast', axis=1, inplace=True)

    df=df.reset_index()
    return df

```

# Getting Infos

```python
# Create a dictionary of dataframes so save all the infos

d={}
for year in urls.year:
    d[year]=pd.DataFrame()

d1={}
for year in urls.year:
    d1[year]=pd.DataFrame()
    
d2={}
for year in urls.year:
    d2[year]=pd.DataFrame()
```

```python
#getting infos from the list's first page
for i in range(23):
    d[urls['year'][i]]=correct(imdb_100(urls['link'][i], 100), 100)
```

```python
#getting infos from the list's second page
for i in range(23):
    d1[urls['year'][i]]=correct(imdb_100(urls['page2'][i], 100), 100)
```

```python
#getting infos from the list's first page
for i in range(23):
    d2[urls['year'][i]]=correct(imdb_100(urls['page3'][i], 50), 50)
```

```python
#Concatenating all dictionaries of dataframes
df=pd.concat(d, ignore_index=True)
df1=pd.concat(d1, ignore_index=True)
df2=pd.concat(d2, ignore_index=True)

dfs=pd.concat([df, df1, df2], ignore_index=True)

dfs['IMDByear']=dfs['IMDByear'].apply(lambda x: x[13:])
dfs['Date']=dfs['Date'].astype(str)
dfs['Date']=dfs['Date'].apply(lambda x: x.replace('II ', '').replace('I ', '').replace(' TV Movie', '').replace(' TV Short', ''))
dfs=dfs.sort_values(['IMDByear', 'Ranking'])
```

# Getting Top1000 Voter's Values

```python
dff=dfs[['Ranking', 'IMDByear', 'Title', 'IMDBlink']]
dff['IMDBlink']=dff['IMDBlink'].apply(lambda x: "https://www.imdb.com/" + x + "ratings?demo=top_1000_voters")
dff=dff.drop_duplicates('IMDBlink')
dff=dff.reset_index()
dff=dff.drop('index', axis=1)
dff
```

```python
def topvote(link):
        page=get(link)
        soup = BeautifulSoup(page.content, 'lxml')
        f = soup.find_all("div", class_='allText')

        top=f[0].text.split('\n')[1].strip()
        return top
```

```python
dff['#top1000']=dff['IMDBlink'].apply(lambda x: topvote(x))
```

```python
def vote_res(link):
    
    page=get(link)
    soup = BeautifulSoup(page.content, 'lxml')
    f = soup.find_all("div", class_='allText')
    r10 = f[3].text.strip()
    r9 = f[6].text.strip()
    r8 = f[9].text.strip()
    r7 = f[12].text.strip()
    r6 = f[15].text.strip()
    r5 = f[18].text.strip()
    r4 = f[21].text.strip()
    r3 = f[24].text.strip()
    r2 = f[27].text.strip()        
    r1 = f[30].text.strip()
    
    rates=[r10, r9, r8, r7, r6, r5, r4, r3, r2, r1]
    return rates
```

```python
dff['Rates']=dff['IMDBlink'].apply(lambda x: vote_res(x))
```
